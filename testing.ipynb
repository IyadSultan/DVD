```python
import pandas as pd
from IPython.display import display, HTML

# Read the CSV file
df = pd.read_csv('samples/discharge_samples_200.csv', encoding='latin1')

# Add word count column
df['word_count'] = df.iloc[:, 1].str.split().str.len()

# Function to display a specific note
def display_note(note_number):
    if 0 <= note_number < len(df):
        note = df.iloc[note_number]
        html = f"""
        <div style='background-color: #f9f9f9; padding: 10px; border-radius: 5px; margin: 10px 0; color: #000000;'>
            <h3 style='color: #000000;'>Note #{note.iloc[0]}</h3>
            <p style='color: #000000;'><b>Word count:</b> {note['word_count']}</p>
            <p style='color: #000000;'><b>Note text:</b></p>
            <p style='color: #000000; white-space: pre-wrap;'>{note.iloc[1]}</p>
        </div>
        """
        display(HTML(html))
    else:
        print(f"Note number must be between 0 and {len(df)-1}")

# Example usage:
print(f"Total notes available: {len(df)}")
print("To view a note, run: display_note(note_number)")
print("For example: display_note(0) to see the first note")

# # Display the first note as an example
# display_note(3)
```
